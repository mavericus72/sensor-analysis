```python
import numpy as np
```

```python
data = np.genfromtxt(
    "../data/sensor_data.csv",
    delimiter=",",
    skip_header=1
)
```

```python
print(data.shape)
```

```python
data[:5]
```

```python
rpm = data[:, 0]
temperature = data[:, 1]
vibration = data[:, 2]
tyre_pressure = data[:, 3]
```

```python
print("RPM:", rpm[:5])
print("Temperature:", temperature[:5])
print("Vibration:", vibration[:5])
print("Tyre Pressure:", tyre_pressure[:5])
```

```python
import matplotlib.pyplot as plt
```

```python
time = range(len(rpm))
```

```python
plt.figure()
plt.plot(time, rpm)
plt.xlabel("Time")
plt.ylabel("RPM")
plt.title("RPM Sensor Over Time")
plt.show()
```

```python
plt.figure()
plt.plot(time, temperature)
plt.xlabel("Time")
plt.ylabel("Temperature")
plt.title("Engine Temperature Over Time")
plt.show()
```

```python
plt.figure()
plt.plot(time, vibration)
plt.xlabel("Time")
plt.ylabel("Vibration")
plt.title("Vibration Sensor Over Time")
plt.show()
```

```python
plt.figure()
plt.plot(time, tyre_pressure)
plt.xlabel("Time")
plt.ylabel("Tyre Pressure")
plt.title("Tyre Pressure Over Time")
plt.show()
```

```python
print("RPM Mean:", rpm.mean())
print("RPM Std Dev:", rpm.std())

print("Temperature Mean:", temperature.mean())
print("Temperature Std Dev:", temperature.std())

print("Vibration Mean:", vibration.mean())
print("Vibration Std Dev:", vibration.std())

print("Tyre Pressure Mean:", tyre_pressure.mean())
print("Tyre Pressure Std Dev:", tyre_pressure.std())
```
